Handling Outliers

    Investigation: Determine if outliers are due to data entry errors or if they represent true variability in the data.

    Transformation: Apply transformations (e.g., log transformation) to reduce the impact of outliers.

Q7. You are a data analyst for a flight booking website, and you have been asked to analyze the Flight
Price dataset to identify any trends in flight prices. What features would you analyze to identify these
trends, and what visualizations would you use to present your findings to your team?


To identify trends in flight prices, you can focus on the following features:

    Date_of_Journey: Analyze the prices based on the date of travel to identify seasonal trends. 

    Airline: Compare prices across different airlines to identify brand-specific pricing trends.

    Source and Destination: Examine how prices vary by departure and arrival locations.

    Duration: Look at how flight duration affects prices.

    Total_Stops: Analyze the impact of the number of stops on flight prices.

    Price: This is the target variable you will be analyzing for trends.

```python

df['Date']= df['Date_of_Journey'].str.split('/').str[0].astype(int)
df['Month']= df['Date_of_Journey'].str.split('/').str[1].astype(int)
df['year']= df['Date_of_Journey'].str.split('/').str[2].astype(int)

# Analyze trends by month,year and date
monthly_avg_price = df.groupby('Month')['Price'].mean()
yearly_avg_price = df.groupby('year')['Price'].mean()
date_avg_price = df.groupby('Date')['Price'].mean()

# Plot the monthly average price
plt.figure(figsize=(12, 6))
monthly_avg_price.plot(kind='line', marker='o', color='b')
plt.title('Average Flight Price by Month')
plt.xlabel('Month')
plt.ylabel('Average Price')
plt.grid(True)
plt.show()


# Analyze trends by airline
airline_avg_price = df.groupby('Airline')['Price'].mean()

# Plot the average price by airline
plt.figure(figsize=(12, 6))
airline_avg_price.plot(kind='bar', color='m')
plt.title('Average Flight Price by Airline')
plt.xlabel('Airline')
plt.ylabel('Average Price')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

# Analyze trends by source and destination
source_avg_price = df.groupby('Source')['Price'].mean()
destination_avg_price = df.groupby('Destination')['Price'].mean()

# Plot the average price by source
plt.figure(figsize=(12, 6))
source_avg_price.plot(kind='bar', color='orange')
plt.title('Average Flight Price by Source')
plt.xlabel('Source')
plt.ylabel('Average Price')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

# Plot the average price by destination
plt.figure(figsize=(12, 6))
destination_avg_price.plot(kind='bar', color='purple')
plt.title('Average Flight Price by Destination')
plt.xlabel('Destination')
plt.ylabel('Average Price')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

# Analyze trends by number of stops
stops_avg_price = df.groupby('Total_Stops')['Price'].mean()

# Plot the average price by number of stops
plt.figure(figsize=(12, 6))
stops_avg_price.plot(kind='bar', color='teal')
plt.title('Average Flight Price by Number of Stops')
plt.xlabel('Total Stops')
plt.ylabel('Average Price')
plt.xticks(rotation=0)
plt.grid(True)
plt.show()
```

# Insight 

Monthly Trends: Flight prices peak in march , may and june ,in april  flight prices are low

Airline Comparison: Jet Airways Business airline having higher prices than other airlines

Source and Destination: Delhi have higher average prices, then Kolkata and then Banglore 

Number of Stops: 4 stops flight tend to be more expensive compared to flights with other stops.

```python
df_copy["Reviews"] = df_copy["Reviews"].astype(int)
```

```python
df_copy.info()
```

```python

# Group by 'Category' and calculate total installs, average rating, and total reviews
category_analysis = df_copy.groupby('Category').agg({
    'Installs': 'sum',
    'Rating': 'mean',
    'Reviews': 'sum'
}).reset_index()

# Sort by total installs to identify the most popular categories
category_analysis = category_analysis.sort_values(by='Installs', ascending=False)

# Display the top categories based on installs
print(category_analysis.head(10))

# Visualize the data: Top 10 categories by total installs
plt.figure(figsize=(12, 8))
plt.barh(category_analysis['Category'].head(10), category_analysis['Installs'].head(10), color='skyblue')
plt.xlabel('Total Installs')
plt.title('Top 10 App Categories by Total Installs')
plt.gca().invert_yaxis()
plt.show()

# Visualize the data: Top 10 categories by average rating
top_10_avg_rating = category_analysis.sort_values(by='Rating', ascending=False).head(10)
plt.figure(figsize=(12, 8))
plt.barh(top_10_avg_rating['Category'], top_10_avg_rating['Rating'], color='lightgreen')
plt.xlabel('Average Rating')
plt.title('Top 10 App Categories by Average Rating')
plt.gca().invert_yaxis()
plt.show()
```
